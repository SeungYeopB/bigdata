## 박스 플롯

![boxplot](data/lec8_Visualization/boxplot.png) 

* 가운데 부분을 Box라고 부르고,  양쪽 끝을 콧수염을 닮아 whisker 라고 부르며
* box and whisker plot 이라고도 칭한다.
* 바깥의 점들은 이상점(outliers) 이다.  ✫ outliers: 평균값들과 너무 달라 통계에 방해가 될 수 있는 요소들

### 표 분석
1) whisker 범위가 넓다 : 값들이 다양하게 분포되어있다.
2) y값이 다른 boxplot들 보다 비교적 높다:  비교적 높은 숫자들이 분포해 있다.
3) 박스의 크기, 박스부터 whisker까지의 길이 : 그 부분의 값들의 분산값을 나타낸다.
